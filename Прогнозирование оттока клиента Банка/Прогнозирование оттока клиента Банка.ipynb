Видим, что распределение далеко от нормального, значения варьируют от 0 до 10 лет со средним и медианным значениями в 5 лет. 

Заполним пропущенные значения случайными величинами и добавим в таблицу дополнительную колонку с меткой о пропуске `missed_data`, где `False` укажет на истинные значения, а `True` на синтетические.

```python
df['missed_data'] = df.tenure.isna().astype('int')
```

Напишем формулу для заполнения случайными значениями пропусков в колонке.

```python
def random_tenure(data):
    if data['missed_data'] == 1:
        return random.randint(0, 10)
    return data['tenure']
```

Применим формулу для заполнения пропусков.

```python
df['tenure'] = df.apply(random_tenure, axis=1)
```

Оценим, изменилось ли распределение.

```python
df.tenure.hist(bins=21)
plt.title('Сколько лет человек является клиентом банка')
plt.ylabel('Количество клиентов')
plt.xlabel('Количество лет')
plt.show()
```

Видим, что распределение осталось тем же.

## Исследование задачи

Существует явный дисбаланс в распределении целевого признака. Однако в качестве стартовой точки посмотрим, как обучится модель без учета дисбаланса классов.

### Стартовая модель логистической регрессии

Подсчитаем стартовую модель логистической регрессии.

```python
start_model_regression = LogisticRegression(random_state=state, solver='liblinear')
start_model_regression.fit(features_train, target_train)
predictions_valid = start_model_regression.predict(features_valid)

start_regression_precision = precision_score(target_valid, predictions_valid)
start_regression_recall = recall_score(target_valid, predictions_valid)

start_regression_f1 = f1_score(target_valid, predictions_valid)

print("Полнота:", start_regression_recall)
print("Точность:", start_regression_precision)
print("F1-мера:", start_regression_f1)
```

Полнота и точность оценивают качество прогноза положительного класса с разных позиций. Recall описывает, как хорошо модель разобралась в особенностях этого класса и распознала его. Precision выявляет, не переусердствует ли модель, присваивая положительные метки.

Видим, что пока метрики далеки от желаемых (не менее `0.59` для метрики F1).

Посмотрим на значения метрики AUC-ROC. Для случайной модели значения данной метрики равны 0.5, соответственно у нас она должна оказаться выше. В отличие от других метрик, на вход она принимает не предсказания, а
вероятности класса «1», поэтому необходимо рассчитать их. 

Создадим переменную `probabilities_one_valid` для предсказаний положительного класса.

```python
probabilities_one_valid = start_model_regression.predict_proba(features_valid)[:, 1]
```

### Стартовая модель решающего дерева
<a id='best_tree'> </a>

Посчитаем стартовую модель решающего дерева.

```python
%%time
start_model_tree = None
start_model_tree_depth = 0
start_tree_precision = 0
start_tree_recall = 0
start_tree_f1 = 0

for depth in range(1, 30):
    start_model_tree_test = DecisionTreeClassifier(
        random_state=state,
        max_depth = depth
    )
    start_model_tree_test.fit(features_train, target_train)
    predictions_valid = start_model_tree_test.predict(features_valid)
    tree_f1 = f1_score(target_valid, predictions_valid)
    if tree_f1 > start_tree_f1:
        start_tree_f1 = tree_f1
        start_model_tree = start_model_tree_test
        start_model_tree_depth = depth
        start_tree_precision = precision_score(target_valid, predictions_valid)
        start_tree_recall = recall_score(target_valid, predictions_valid)

print("Полнота:", start_tree_recall)
print("Точность:", start_tree_precision)
print("F1-мера:", start_tree_f1)
print('Глубина дерева:', start_model_tree_depth)
```

Значение метрики F1 уже ближе к желаемой.

Посмотрим на значение метрики AUC-ROC.

```python
probabilities_one_valid_tree_start = start_model_tree.predict_proba(features_valid)[:, 1]
auc_roc_start_tree = roc_auc_score(target_valid, probabilities_one_valid_tree_start)
auc_roc_start_tree
```

Значения метрики выше, чем у логистической регрессии.

### Стартовая модель случайного леса

Посчитаем стартовую модель случайного леса.

Начнем с того, что найдем подходящую глубину дерева.

```python
%%time
start_model_forest = None
start_model_forest_depth = 0
start_model_forest_est = 20

start_forest_precision = 0
start_forest_recall = 0
start_forest_f1 = 0

for depth in range(1, 30):
    start_model_forest_test = RandomForestClassifier(
        random_state=state,
        max_depth=depth,
        n_estimators=start_model_forest_est
    )
    start_model_forest_test.fit(features_train, target_train)
    predictions_valid = start_model_forest_test.predict(features_valid)
    forest_f1 = f1_score(target_valid, predictions_valid)
    if forest_f1 > start_forest_f1:
        start_forest_f1 = forest_f1
        start_model_forest = start_model_forest_test
        start_model_forest_depth = depth
        start_forest_precision = precision_score(target_valid, predictions_valid)
        start_forest_recall = recall_score(target_valid, predictions_valid)

print("Полнота:", start_forest_recall)
print("Точность:", start_forest_precision)
print("F1-мера:", start_forest_f1)
print('Глубина дерева:', start_model_forest_depth)
```

Нашли подходящую глубину и повторим то же самое для нахождения подходящего количества деревьев в модели.

### Увеличение выборки

Когда обучают модели, чтобы сбалансировать классы путём увеличения числа, такая техника называется `upsampling`.

Преобразование проходит в несколько этапов:
1. Разделить обучающую выборку на объекты по классам;
2. Определить тот класс, который содержит меньше объектов. Назовём его меньшим классом;
3. Скопировать несколько раз объекты меньшего класса;
4. С учётом полученных данных создать новую обучающую выборку;
5. Перемешать данные.

[Соотношение классов в таблице 1:4.](#exited)
Можем увеличить выборку меньшего класса в 4 раза.

Напишем формулу для увеличения выборки.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)

    return features_upsampled, target_upsampled
```

Увеличим меньший класс тренировочной выборки.

### Уменьшение выборки

Вместо повторения объектов меньшего класса, уберём часть обектов большего класса. Это можно сделать техникой `downsampling`.

Преобразование проходит в несколько этапов:
1. Разделить обучающую выборку на объекты по классам;
2. Определить тот класс, который содержит больше объектов. Назовём его большим классом;
3. Случайным образом отбросить часть из объектов большего класса;
4. С учётом полученных данных создать новую обучающую выборку;
5. Перемешать данные.

[Соотношение классов в таблице 1:4.](#exited)
Можем уменьшить выборку большего класса в 4 раза. 

Напишем функцию для уменьшения выборки

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [
            features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [
            target_ones])

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)

    return features_downsampled, target_downsampled
```

## Изменение порога

Мы убедились, что для логистической регрессии и решающего дерева ни стартовая модель, ни один из методов борьбы с дисбалансом не достигли метрики F1 в 0.59. Поэтому дальше мы рассмотрим, как повлияет изменение порога для метрику F1.

Для изменения порога мы возьмем модели, показавшие максимальные значения метрики F1.

Таким образом, для каждой модели выберем следующие методы борьбы с дисбалансом:
1. Для **логистической регрессии** - взвешивание классов, модель [`balanced_model_regression`](#best_regr).
2. Для **решающего дерева** - стартовая модель [`start_model_tree`](#best_tree).
3. Для **случайного леса** - взвешивание класса [`balanced_model_forest`](#best_forest).

### Модель логистической регрессии

Напишем функцию для подбора порога, при котором метрика качества F1 будет иметь максимальное значение.

```python
def threshold_find(start, finish, step, probabilities_one, target, show=True):
    '''функция перебирает указанные элементы диапазона для порога и возращает лучшее значение метрики F1
    start, finish, step - начало, конец и шаг для перебора порога
    probabilities_one_valid - вероятности положительного класса
    target - значения целевого признака
    show - выводить ли на экран значения всех метрик, по умолчанию - выводить'''
    
    f1 = 0
    precision = 0
    recall = 0
    threshold = None
    for step_threshold in np.arange(start, finish, step):
        
        predicted = probabilities_one > step_threshold
        step_f1 = f1_score(target, predicted)
        
        if step_f1 > f1:
            f1 = step_f1
            precision = precision_score(target, predicted)
            recall = recall_score(target, predicted)
            threshold = step_threshold
    if show == True:
        print("Порог = {:.2f} | Полнота = {:.3f}, Точность = {:.3f}".format(
            threshold, recall, precision))
        print('F1: {:.3f}'.format(f1))

    return recall, precision, f1, threshold
```

Подберем наиболее подходящий порог для модели [логистической регрессии со взвешенными классами](#best_regr).

```python
threshold_regression_recall, threshold_regression_precision, threshold_regression_f1, threshold_best_regression = threshold_find(
    0.1, 
    0.99, 
    0.01, 
    balanced_regression_probabilities_one_valid,
    target_valid
)
```

Так как графики однотипные, упростим себе визуализацию, записав повторяющиеся команды в функцию.

```python
def pr_show(recall, precision, x_recall=None, y_precision=None, line_show=True, add_title=None, add_figsize=(12, 4)):
    plt.figure(figsize=add_figsize)
    
    plt.step(recall, precision, where='post')
    
    plt.xlabel('Recall', color='green')
    plt.ylabel('Precision', color='b')
    
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    
    plt.title(add_title)
    
    if line_show == True:
        plt.axvline(x=x_recall, color='g', linestyle='--')
        plt.axhline(y=y_precision, color='b', linestyle='--')
    
    plt.show()
```
